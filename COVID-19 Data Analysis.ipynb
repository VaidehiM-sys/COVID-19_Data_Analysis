# COVID-19 Data Analysis


Exploratory Data Analysis performed on COVID-19 cases from January 2020 to April 2020 identified in US.

The data is from the website:
https://usafacts.org/visualizations/coronavirus-covid-19-spread-map/

There are three files:

• covid_deaths_usafacts.csv

• covid_confirmed_usafacts.csv

• covid_county_population_usafacts.csv

### covid_deaths_usafacts.csv

This contains the known deaths in each county in the USA each day from January 22 to April 6. 

The column headers besides the dates are:

• countyFIPS - Federally assigned number identifying each county in the country

• County Name - obvious

• State - The state the county is in

• stateFIPS - Federally assigned number identifying each state in the country. This number starts the countyFIPS


### covid_confirmed_usafacts.csv

This file contains the know covid cases in each county in the USA each day from January 22 to
April 5. The column headers are the same as above.

### covid_county_population_usafacts.csv

This file contains the population of each county in the USA. The column headers are:

• countyFIPS - Federally assigned number identifying each county in the country

• County Name

• State - The state the county is in

• population - Population of the county

### Importing the required packages

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

### Reading the data from the CSV files

```python
# The file directory storing all the required file
os.chdir(r"..\covid-19")
directory = os.getcwd()
print("The directory to read the files from is: ", directory)
```

```python
# Concatinating the absolute file path and file name
covid_confirmed_file = os.path.join(os.path.abspath(directory), 'covid_confirmed_usafacts.csv')
covid_county_population_file = os.path.join(os.path.abspath(directory), 'covid_county_population_usafacts.csv')
covid_deaths_file = os.path.join(os.path.abspath(directory), 'covid_deaths_usafacts.csv')

if (os.path.exists(covid_confirmed_file)):
    # Reading the COVID confirmed usafacts CSV file data into data frame
    covid_confirmed_cases = pd.read_csv(covid_confirmed_file, sep=',')
else:
    print('covid_confirmed_usafacts file does not exists at the location')
    
if (os.path.exists(covid_county_population_file)):
    # Reading the COVID county population usafacts CSV file data into data frame
    covid_county_population_cases = pd.read_csv(covid_county_population_file, sep=',')
else:
    print('covid_county_population_usafact file does not exists at the location')
    
    
if (os.path.exists(covid_deaths_file)):
    # Reading the COVID deaths usafacts CSV file data into data frame
    covid_deaths = pd.read_csv(covid_deaths_file, sep=',')
else:
    print('covid_deaths_usafacts file does not exists at the location')    
```

### Task 1: Compute the total confirmed cases per day in the country. Produce a table and a line plot.


```python
# Displaying first 5 records in covid_confirmed_cases dataframe
covid_confirmed_cases.head()
```

```python
covid_confirmed_cases.describe()
```

### Task 2 : Compute the total confirmed cases per week in the country. Plot the results using a log line plot. That is the log of the number of cases. A log plot of an exponential process will produce a straight line.

```python
confirmed_cases_per_week = confirmed_cases_per_day.copy()
```

```python
#Computing the Week Number from the date and grouping by Week to take the last row 
confirmed_cases_per_week['Date'] = pd.to_datetime(confirmed_cases_per_week.index)
confirmed_cases_per_week['Week'] = confirmed_cases_per_week['Date'].dt.week
confirmed_cases_per_week = confirmed_cases_per_week.groupby('Week').tail(1).drop(['Date'],axis=1)
confirmed_cases_per_week = confirmed_cases_per_week[['Week','ConfirmedCases']].reset_index(drop=True)
confirmed_cases_per_week
```

```python
# Displaying the plot for the number of cases first to compare two graphs 
figure(figsize=(15, 5), facecolor='w', edgecolor='k')
plt.style.use('ggplot')
plt.plot(confirmed_cases_per_week['Week'],  confirmed_cases_per_week['ConfirmedCases'])
plt.xlabel('Week')
plt.ylabel('Number of confirmed cases')
plt.title('COVID 19 confirmed cases per Week from 22 January 2020 to 6 April 2020')
plt.tight_layout()
plt.show()
```

```python
# Displaying the plot for the log of cases 
figure(figsize=(15, 5), facecolor='w', edgecolor='k')
plt.style.use('ggplot')
plt.plot(confirmed_cases_per_week['Week'], np.log(confirmed_cases_per_week['ConfirmedCases']))
plt.xlabel('Week')
plt.ylabel('Log of number of confirmed cases')
plt.title('Log of COVID 19 confirmed cases per Week from 22 January 2020 to 6 April 2020')
plt.tight_layout()
plt.show()
```

### The resulting log plot  of Week VS log(Cases) looks exponential

### Task 5: Find the 10 counties with the most known number of covid-19 cases. Plot the number of cases each week.

```python
county_covid_confirmed_cases = pd.DataFrame(covid_confirmed_cases.groupby('County Name', sort=True)['4/6/2020'].sum().reset_index())
top_10_county_with_cases = county_covid_confirmed_cases.rename(columns={'4/6/2020':'TotalCases'}).sort_values(by='TotalCases',ascending=False).head(10)
```

```python
top_10_county_with_cases.reset_index(drop=True)
```

```python
# Finding the top cases 
top_10_cases = covid_confirmed_cases.nlargest(10, ['4/6/2020'])
top_10_cases.drop(['countyFIPS', 'State', 'stateFIPS'], axis=1, inplace=True)
top_10_cases = top_10_cases.set_index('County Name')
top_10_cases 
```

```python
#Transposing the matrix 
top_10_cases = top_10_cases.T
top_10_cases['Date'] = pd.Series(index=top_10_cases.index, dtype='datetime64[ns]')
top_10_cases['Date'] = pd.to_datetime(top_10_cases.index)
top_10_cases['Week'] = top_10_cases['Date'].dt.week
top_10_cases.drop(['Date'], axis=1, inplace=True)
top_10_cases = top_10_cases.groupby(['Week']).tail(1).set_index('Week')
top_10_cases
```

```python
ax = top_10_cases.plot(kind='bar', title ="Weekly Cases in Counties", figsize=(15, 10), legend=True, fontsize=12, logy=True)
ax.set_xlabel("Cases", fontsize=12)
ax.set_ylabel("Week", fontsize=12)
plt.show()
```

#### Similarities and Differences
#### In all counties cases increased from week 10 onwards
#### Only Cook County had cases from week 4 to week 9 

### Task 6: Compute the number of known cases per population in each county that have cases of covid-19. Produce a bar graph and table of the results for the 10 counties with the highest values and the 10 counties with the lowest values.

```python
cases_by_population = pd.concat([covid_county_population_cases, covid_confirmed_cases], axis=1)
cases_by_population = cases_by_population.loc[:, ~cases_by_population.columns.duplicated()]
cases_by_population['CasesPerPopulation'] = cases_by_population['4/6/2020']*100000/cases_by_population['population']
cases_by_population = cases_by_population.replace([np.inf, -np.inf], np.nan).dropna()
cases_by_population
```

```python
# Top 10 counties with lowest values
lowest_cases = cases_by_population[cases_by_population['CasesPerPopulation'] > 0]
lowest_cases = lowest_cases.sort_values('CasesPerPopulation', ascending=True).head(10)
lowest_cases
```

```python
# Top 10 counties with highest values
highest_cases = cases_by_population[cases_by_population['CasesPerPopulation'] > 0]
highest_cases = highest_cases.sort_values('CasesPerPopulation', ascending=False).head(10)
highest_cases
```

```python
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.style.use('ggplot')
plt.bar(lowest_cases['County Name'], lowest_cases['CasesPerPopulation'], linestyle='solid')
plt.bar(highest_cases['County Name'], highest_cases['CasesPerPopulation'], linestyle='solid')
plt.yscale('log')
plt.xlabel('County Name')
plt.ylabel('Covid Cases')
plt.title('Cases per 100000 population')
plt.gcf().autofmt_xdate(rotation=45)
plt.show()
```

### Task 7: Compute the death rate (deaths/ known cases) in each county that have both deaths and known cases. Compute the mean and standard deviation of the results. 

```python
deaths = pd.DataFrame()
deaths['CountyName'] = covid_confirmed_cases['County Name']
deaths['CovidCases'] = covid_confirmed_cases['4/6/2020']
deaths['TotalDeaths'] = covid_deaths['4/6/2020']
deaths
```

```python
deaths = deaths[(deaths['CovidCases'] > 0) & (deaths['TotalDeaths'] > 0)]
deaths['DeathRate'] = deaths['TotalDeaths']/deaths['CovidCases']
deaths['CovidCases'] = deaths['CovidCases'].astype('int')
deaths['TotalDeaths'] = deaths['TotalDeaths'].astype('int')
deaths
```

```python
print('Mean Death Rate is :', deaths.DeathRate.mean())
```

```python
print('Standard Deviation of Death Rate is :', deaths.DeathRate.std())
```

### Task 8: Produce a violin plot of the death rates found in #7 

```python
# Producing the violin plot for death rates
fig, ax = plt.subplots(figsize =(9, 7)) 
sns.violinplot( ax = ax, y = deaths['DeathRate'] ) 
```

### Task 9: There is some hope that covid-19 may have peaked in NYC (counties Bronx, Kings, New York, Queens and Richmond). Is there any evidence of that in the data?

```python
ny_covid_cases = covid_confirmed_cases.loc[covid_confirmed_cases['State'] == 'NY'] 
county_list = ['Bronx County','Kings County', 'New York County', 'Queens County', 'Richmond County']
ny_covid_cases = ny_covid_cases[ny_covid_cases['County Name'].isin(county_list)]
ny_covid_cases = ny_covid_cases.drop(['countyFIPS', 'State', 'stateFIPS'], axis=1)
ny_covid_cases = ny_covid_cases.set_index('County Name')
```

```python
ny_covid_cases[ny_covid_cases.columns[-1]]
```

```python
ny_covid_cases = ny_covid_cases.T
figure(num=None, figsize=(30, 6), dpi=80, facecolor='w', edgecolor='k')
plt.style.use('fivethirtyeight')
plt.plot(ny_covid_cases.tail(40), linestyle='solid')
plt.title('Cases Per Counties')
plt.xlabel('Counties')
plt.ylabel('Cases')
plt.yscale('log')
plt.gcf().autofmt_xdate(rotation=90)
plt.legend(ny_covid_cases.columns, loc='upper left')
plt.tight_layout()
plt.show()
```

### COVID 19 is peaked in NYC as per above graph 

### Task 10: Seattle (King county in Washington) also hopes that they have peaked. Is there any evidence of that in the data?

```python
seattle_cases = covid_confirmed_cases[(covid_confirmed_cases['County Name']=='King County') & (covid_confirmed_cases['State']=='WA')]
seattle_cases = seattle_cases.set_index('County Name')
seattle_cases = seattle_cases.drop(['countyFIPS', 'State', 'stateFIPS'], axis=1)
seattle_cases = seattle_cases.T
seattle_cases
```

```python
figure(num=None, figsize=(30, 6), dpi=80, facecolor='w', edgecolor='k')
plt.style.use('ggplot')
ax = sns.heatmap(linewidth=0.5, data=seattle_cases.T)
plt.tight_layout()
plt.show()
```

### The above heatmap shows that cases in Seattle have peaked.
